```python
#Applying sharpening filter Lena.png

import cv2
import numpy as np

image = io_url.imread('https://github.com/arwajlokhandwala/467Module2LabAssignment/blob/main/Lena.png?raw=true')
image = np.float32(image)/255

kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])
sharpened_image = cv2.filter2D(image,-1,kernel)
plt.imshow(sharpened_image)
```
